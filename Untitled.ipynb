```python
import pandas as pd
```

```python
df=pd.read_excel('amazon_products.xlsx')
```

```python
df
```

```python
unnamed_columns = [col for col in df.columns if col.startswith('Unnamed:')]

# Delete unnamed columns
for col in unnamed_columns:
    del df[col]
```

```python
# Convert 'title' column to string data type
df['title'] = df['title'].astype(str)
```

```python
df['isBestSeller']=df['isBestSeller'].astype(bool)
```

```python
df['stars'] = df['stars'].str.replace("2024", "2023")
```

```python
df['stars'] = df['stars'].str.slice(stop=10)
```

```python
df['stars'] = pd.to_numeric(df['stars'], errors='coerce', downcast='integer')
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce', downcast='integer')
```

```python
df['listPrice'] = pd.to_numeric(df['listPrice'], errors='coerce', downcast='integer')
```

```python
df['boughtInLastMonth'] = pd.to_numeric(df['boughtInLastMonth'], errors='coerce', downcast='integer')
```

```python
df
```
